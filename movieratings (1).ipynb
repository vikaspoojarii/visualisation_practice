```python
listl = []  
mylabels = []  

for gen in movies.Genre.cat.categories:
    listl.append(movies[movies.Genre == gen]["Budgetmillion"])  
    mylabels.append(gen)

sns.set_style("whitegrid")  

fig, ax = plt.subplots(figsize=(11.7, 8.27))  
# Plot the histogram
plt.hist(listl, bins=30, stacked=True, rwidth=1, label=mylabels)  

# Adding title and labels
plt.title("Movie Budget Distribution", fontsize=35, color="DarkBlue",) 
plt.ylabel("Number of Movies", fontsize=25, color="Red")  
plt.xlabel("Budget (million)", fontsize=25, color="Green") 

# Formatting ticks
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)

# Adding legend
plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=1, prop={'size': 20}) 

plt.show()  
```

Movie Ratings Analysis and Insights
This project explores the relationship between movie budgets, audience ratings, and critic ratings across various genres. The goal is to uncover trends and patterns that can provide actionable insights for stakeholders in the film industry.

Visualizations and Key Insights
1. Budget Distribution by Genre (Stacked Histogram)
Insight: Movies with budgets below $50 million dominate most genres. Action and Drama movies appear more frequently across all budget categories.
Conclusion: Low-budget movies are prevalent, but higher-budget Action and Drama films may drive audience engagement due to their frequency.
2. Critic Rating vs Audience Rating by Genre and Year (FacetGrid)
Insight: Audience and critic ratings vary significantly by genre and year. Comedy and Drama are more consistent in ratings, while genres like Adventure and Horror show greater variation.
Conclusion: Rating trends highlight the need to adapt strategies based on evolving audience preferences for specific genres.
3. Critic Ratings by Genre (Box Plot)
Insight: Drama and Thriller movies exhibit the widest range of critic ratings, indicating a mix of highly rated and poorly rated films. Comedy and Adventure genres tend to have more consistent critic ratings.
Conclusion: These variations suggest that certain genres are more polarizing for critics, potentially influencing marketing strategies.
4. Audience Ratings vs Critic Ratings (Scatterplot)
Insight: There is a strong positive correlation between audience and critic ratings. However, there are noticeable outliers where movies are favored by one group over the other.
Conclusion: Films should be marketed differently depending on whether they aim to appeal to critics, audiences, or both.


Tools Used
Python Libraries: Pandas, Seaborn, Matplotlib
Visualization Techniques: Scatterplots, Box Plots, Stacked Histograms, FacetGrids
Key Takeaways
This analysis highlights the importance of understanding audience preferences by genre and budget. Insights from critic and audience ratings can help producers and marketers make informed decisions about future movie projects.
